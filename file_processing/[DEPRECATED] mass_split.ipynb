```python
import pandas as pd
import logging
import json
from datetime import datetime
import os
from splitter import separate_track
data_path = 'DownloadViaChannel.xlsx'
df = pd.read_excel(data_path)
```

```python
#  # Set up logging
# logging.basicConfig(
#     level=logging.INFO,
#     format='%(asctime)s - %(levelname)s - %(message)s',
#     handlers=[
#         logging.FileHandler("file_processing.log", encoding="utf-8"),
#         logging.StreamHandler()  # Also print to console
#     ]
# )
# Load or create tracking file
tracking_file = "processed_files.json"
try:
    with open(tracking_file, 'r') as f:
        processed_files = json.load(f)
except (FileNotFoundError, json.JSONDecodeError):
    processed_files = {}
```

```python
# # Your processing code
# for row in df.itertuples():
#     input_dir = os.path.join("audio_output_viachannel", row.company, str(row.generation), row.group)
#     output_dir = os.path.join("separated_stems", row.company, str(row.generation), row.group)
    
#     logging.info(f"Processing group: {row.company}/{row.generation}/{row.group}")
    
#     for input_file in os.listdir(input_dir):
#         file_path = os.path.join(input_dir, input_file)
#         output_file_path = os.path.join(output_dir, input_file)
        
#         if not os.path.exists(output_file_path):
#             logging.info(f"Processing: {file_path}")
#             try:
#                 separate_track(file_path, output_dir)
#                 logging.info(f"Successfully processed: {file_path}")
#             except Exception as e:
#                 logging.error(f"Error processing {file_path}: {str(e)}")
#         else:
#             logging.info(f"Skipping (already exists): {file_path}")


for row in df.itertuples():
    input_dir = os.path.join("audio_output_viachannel", row.company, str(row.generation), row.group) + '/'
    output_dir = os.path.join("separated_stems", row.company, str(row.generation), row.group) + '/'
    
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)
    
    logging.info(f"Processing group: {row.company}/{row.generation}/{row.group}")
    
    for input_file in os.listdir(input_dir):
        input_file_path = os.path.join(input_dir, input_file)
        file_id = f"{row.company}/{row.generation}/{row.group}/{input_file}"
        
        if file_id not in processed_files:
            logging.info(f"Processing: {input_file_path}")
            try:
                separate_track(input_file_path, output_dir)
                
                # Mark as processed
                processed_files[file_id] = {
                    "processed_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
                    "output_dir": output_dir
                }
                
                # Save tracking file after each successful processing
                with open(tracking_file, 'w') as f:
                    json.dump(processed_files, f, indent=2)
                
                logging.info(f"Successfully processed: {input_file_path}")
            except Exception as e:
                logging.error(f"Error processing {input_file_path}: {str(e)}")
        else:
            logging.info(f"Skipping (in tracking file): {input_file_path}")
```
